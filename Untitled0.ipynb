```python
import pandas as pd
import numpy as np
from sklearn import ensemble
from sklearn import metrics
from sklearn.model_selection import train_test_split
import sklearn.preprocessing as prep
```

```python
data =  pd.read_csv('/content/drive/MyDrive/HatariMuslim/TM9/Dataset/american_income_1994.csv')
data.head()
```

```python
def encoder(df, columns):
    for x in columns :
        labelEncoder = prep.LabelEncoder()
        labelEncoder.fit(df[x])
        df[(x + "_code")] = labelEncoder.transform(df[x])
        df = df.drop([x],axis='columns')
    return df
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/HatariMuslim/TM9/Dataset/american_income_1994.csv')
df1.shape
```

```python
df2 = df1.drop_duplicates(keep='first')
```

```python
df2.isnull().sum()
```

```python
df3 = df2.copy()
```

```python
df3.info()
```

```python
df4 = encoder(df3,['workclass','education','marital_status','occupation','relationship','race','sex','native_country'])
df4
```

```python
df = df4.copy()
```

**MODELING**

```python
df['income'] = df['income'].apply(lambda x: ''.join(x).replace(' ', '').replace('.', ''))
```

```python
df.info()
```

```python
X = df.drop(['income', 'fnlwft'], axis='columns')
y = df['income']
```

```python
df['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=90)
```

```python
model = ensemble.RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
report = metrics.classification_report(y_test, y_pred)

print(accuracy)
print(report)
print(model.feature_importances_)
```
